```python
import sys
import importlib.metadata

# Lista de librer√≠as a verificar
libs = [
    "jupyterlab", "scipy", "tsfel", "pandas", "Boruta", 
    "imbalanced-learn", "tqdm", "pyarrow", "scikit-learn", 
    "lightgbm", "unidecode"
]

print("Python:", f"{sys.version_info.major}.{sys.version_info.minor}.{sys.version_info.micro}")

for lib in libs:
    try:
        version = importlib.metadata.version(lib)
    except importlib.metadata.PackageNotFoundError:
        version = "No instalada / desconocida"
    print(f"{lib}: {version}")
```
